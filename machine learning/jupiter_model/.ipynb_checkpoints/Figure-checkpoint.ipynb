```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
from scipy.stats import *
```

```python
catalog=pd.read_csv('trainset.csv')
catalog
```

```python
filter = (catalog['status']==1)
x11=catalog.loc[:,'u'][filter]
filter = (catalog['status']==0)
x21=catalog.loc[:,'u'][filter]
filter = (catalog['status']==-1)
x31=catalog.loc[:,'u'][filter]

filter = (catalog['status']==1)
x12=catalog.loc[:,'e'][filter]
filter = (catalog['status']==0)
x22=catalog.loc[:,'e'][filter]
filter = (catalog['status']==-1)
x32=catalog.loc[:,'e'][filter]

filter = (catalog['status']==1)
x13=catalog.loc[:,'a_p'][filter]
filter = (catalog['status']==0)
x23=catalog.loc[:,'a_p'][filter]
filter = (catalog['status']==-1)
x33=catalog.loc[:,'a_p'][filter]

filter = (catalog['status']==1)
x14=catalog.loc[:,'inc'][filter]
filter = (catalog['status']==0)
x24=catalog.loc[:,'inc'][filter]
filter = (catalog['status']==-1)
x34=catalog.loc[:,'inc'][filter]


fig,axes=plt.subplots(2,2) 

#上一图
sns.distplot(x11,hist=False,ax=axes[0,0])
sns.distplot(x21,hist=False,ax=axes[0,0])
sns.distplot(x31,hist=False,ax=axes[0,0])

#上二图
sns.distplot(x12,hist=False,ax=axes[0,1])
sns.distplot(x22,hist=False,ax=axes[0,1])
sns.distplot(x32,hist=False,ax=axes[0,1])
#下一图
sns.distplot(x13,hist=False,ax=axes[1,0])
sns.distplot(x23,hist=False,ax=axes[1,0])
sns.distplot(x33,hist=False,ax=axes[1,0])
 #下二图
sns.distplot(x14,kde=False,ax=axes[1,1],label='stable')
sns.distplot(x24,kde=False,ax=axes[1,1],label='esacpe')
sns.distplot(x34,kde=False,ax=axes[1,1],label='encounter')

fig.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1)
fig.suptitle('The stability of circumbinary planet with ${m_p}$=${m_J}$',y=1.05)
fig.tight_layout()#避免重叠

plt.savefig('./jupiter transit.png',dpi=500,bbox_inches = 'tight')
plt.show()
```
