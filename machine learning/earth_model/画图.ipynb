```python
#E
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd

catalog=pd.read_csv('earth_trainset.csv')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
from scipy.stats import *

filter = (catalog['status']==1) 
x11=catalog.loc[:,'u'][filter]
filter = (catalog['status']==0) 
x21=catalog.loc[:,'u'][filter]
filter = (catalog['status']==-1) 
x31=catalog.loc[:,'u'][filter]

filter = (catalog['status']==1) 
x12=catalog.loc[:,'e'][filter]
filter = (catalog['status']==0) 
x22=catalog.loc[:,'e'][filter]
filter = (catalog['status']==-1) 
x32=catalog.loc[:,'e'][filter]

filter = (catalog['status']==1) 
x13=catalog.loc[:,'a_p'][filter]
filter = (catalog['status']==0) 
x23=catalog.loc[:,'a_p'][filter]
filter = (catalog['status']==-1) 
x33=catalog.loc[:,'a_p'][filter]

filter = (catalog['status']==1) 
x14=catalog.loc[:,'inc'][filter]
filter = (catalog['status']==0) 
x24=catalog.loc[:,'inc'][filter]
filter = (catalog['status']==-1) 
x34=catalog.loc[:,'inc'][filter]


fig,axes=plt.subplots(2,2) 

#上一图
sns.distplot(x11,hist=False,kde=True,ax=axes[0,0])
sns.distplot(x21,hist=False,kde=True,ax=axes[0,0])
sns.distplot(x31,hist=False,kde=True,ax=axes[0,0])

#上二图
sns.distplot(x12,hist=False,kde=True,ax=axes[0,1])
sns.distplot(x22,hist=False,kde=True,ax=axes[0,1])
sns.distplot(x32,hist=False,kde=True,ax=axes[0,1])
#下一图
sns.distplot(x13,hist=False,kde=True,ax=axes[1,0])
sns.distplot(x23,hist=False,kde=True,ax=axes[1,0])
sns.distplot(x33,hist=False,kde=True,ax=axes[1,0])
 #下二图
sns.distplot(x14,kde=False,ax=axes[1,1],label='stable')
sns.distplot(x24,kde=False,ax=axes[1,1],label='esacpe')
#sns.distplot(x34,kde=False,ax=axes[1,1],label='encounter')

fig.tight_layout()
#fig.legend.texts[0].set_text("${m_p}$=${m_⊕}$")
fig.suptitle('The stability of circumbinary planet with ${m_p}$=${m_⊕}$',y=1.05)
fig.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1)

plt.savefig('./earth transit.png',dpi=500,bbox_inches = 'tight')
plt.show()
```
