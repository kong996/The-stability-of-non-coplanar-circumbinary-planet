```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
from scipy.stats import *
```

```python
def hw99(df, e, u, p):
    a = 1.6 + 5.1*df[e] 
    -2.22*(df[e]*df[e]) + 4.12*df[u] 
    -4.27*df[e]*df[u] - 5.09*(df[u]*df[u]) 
    +4.61*(df[e]*df[e])*(df[u]*df[u])
    if a < df[p]:
        b = 1
    else:
        b = 0    
    return b
```

```python
df1 = pd.read_csv('testset.csv', sep=',')
df1.status[df1['status']==-1]=0
```

```python
df1['hw99'] = df1.apply(hw99, axis=1, args=('e','u','a_p'))
df1
```

```python
df2 = pd.read_csv('results.csv', sep=',')
df2
```
