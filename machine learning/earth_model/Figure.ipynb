```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
from scipy.stats import *
```

```python
catalog = pd.read_csv('results.csv')
```

```python
filter = (catalog['labels']==1)
x11 = catalog.loc[:,'u'][filter]
filter = (catalog['labels']==0)
x21 = catalog.loc[:,'u'][filter]

filter = (catalog['labels']==1)
x12=catalog.loc[:,'e'][filter]
filter = (catalog['labels']==0)
x22=catalog.loc[:,'e'][filter]

filter = (catalog['labels']==1)
x13=catalog.loc[:,'a_p'][filter]
filter = (catalog['labels']==0)
x23=catalog.loc[:,'a_p'][filter]

x14=[0,1]
x24=[0,1]
```

```python
fig, axes = plt.subplots(2,2)
#上一图
sns.distplot(x11,hist=False,kde=True,ax=axes[0,0])
sns.distplot(x21,hist=False,kde=True,ax=axes[0,0])

#上二图
sns.distplot(x12,hist=False,kde=True,ax=axes[0,1])
sns.distplot(x22,hist=False,kde=True,ax=axes[0,1])

#下一图
sns.distplot(x13,hist=False,kde=True,ax=axes[1,0])
sns.distplot(x23,hist=False,kde=True,ax=axes[1,0])


sns.distplot([0], hist=False,ax=axes[1,1], label='stable')
sns.distplot([0], hist=False,ax=axes[1,1], label='unstable')

#fig.tight_layout()
fig.suptitle('The stability of circumbinary planet with ${m_p}$=${m_⊕}$',y=1.05)
#fig.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1)
plt.show()
```
